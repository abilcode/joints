```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt 
import seaborn as sns

import os



```

## train

```python

data = pd.read_csv("../data/train.csv")
data = data.iloc[:,1:]
data.info()
```

```python
sns.heatmap(data.isna())
```

```python
data.isna().sum()/data.shape[0] 
```

```python
cols_var = [i for i in data.columns]
cols_cat = [i for i in cols_var if data[i].dtypes  == 'object']
cols_num = [i for i in cols_var if i not in cols_cat]
```

```python
for c in cols_var:
    print(f"{c} with dtypes of {data[c].dtypes}")
    print(data[c].nunique())
    print(data[c].unique())
    print(f"There are : {data[c].isna().sum()/data.shape[0] }  missing value")
    print("===="*20)
```

```python
sns.countplot(data.iloc[:,-1])
```

```python
sns.heatmap(data[cols_num].corr(),annot=True)
```

```python
tmp = data.copy()
sns.countplot(
    data=tmp,x="damage_grade", hue="wall_material",
)
```

```python
sns.countplot(
    data=tmp,x="damage_grade", hue="wall_binding",
)
```

```python
sns.countplot(
    data=tmp,x="damage_grade", hue="type_of_reinforcement_concrete",
)

del tmp
```
